### Calcular el tiempo de ejecución de un programa

```
# Importar la librería requerida
import time as tm

# Establecer el tiempo inicial de ejecución
ini_tiempo = tm.time()

# Ejemplo de código al que se le va a medir el tiempo de ejecución
factorial = 1
for i in range(1,100):
  factorial *= i
print("factorial: ", factorial)

# Establecer el tiempo final de ejecución
fin_tiempo = tm.time()

# Calcular el tiempo de ejecución
duracion = fin_tiempo - ini_tiempo

# Mostrar resultados
print(f"El código se empezó a ejecutar a las {ini_tiempo} y terminó de ejecutar"
      f"a las {fin_tiempo}. \nLa duración de la ejecución fue de {duracion}.")
```

### Errores semánticos

El computador no detecta errores semánticos (cuando el programa ejecuta, pero no resuelve de manera adecuada el problema que pretende resolver).

```
# Ejemplo de error semántico

# Código para encontrar el producto de dos números

nmbr1 = 5

nmbr2 = 3

# la siguiente línea presenta un error semántico. La fórmula ingresada corresponde a la suma, no al producto de dos números
producto = nmbr1 + nmbr2

print(f"El producto de {nmbr1} por {nmbr2} es {producto}.")
```

### La estructura **try/except**

Esta estructura permite manejar los errores que se presentan durante la ejecución de un código.

Puede consultar más información sobre el manejo de excepciones en:

[Python Built-in Exceptions](https://docs.python.org/3/library/exceptions.html)

[Exception Handling in Python – Get Rid of those Errors!](https://www.analyticsvidhya.com/blog/2020/04/exception-handling-python/?utm_source=feedburner&utm_medium=email&utm_campaign=Feed%3A+AnalyticsVidhya+%28Analytics+Vidhya%29)


```
# Estructura general try/except

try:
  # Bloque de instrucciones que se van a revisar
  pint("Hola mundo")
except Exception:
  # Acción a tomar si de detecta un error
  print("Tenemos problemas. Hay que revisar el contenido del try.")
else:
  # Acción a tomar si no se detecta un error
  print("Todo salió bien.")
finally:
  # Acción a realizar al terminar la estructura try/except.
  print("Llegamos al final del try/except.")
```

```
# Ensaye a correr este código con numero_2= 0 y con numero_2 = 2
try:
    numero_1 = 2
    numero_2 = 1
    print(numero_1/numero_2)
    print("Programa sin errores")
except ZeroDivisionError:
    print("Para dividir, el numero 2 no puede ser 0")
except TypeError:
    print("Debe ingresar un numero")
except Exception:
    print("Controlo cualquier tipo de error")
finally:
    print("Fin calculadora")
```

### Raise

Genera una excepción cuando se presenta alguna condición especial durante la ejecución del programa. Suspende la ejecución del programa y muestra un mensaje de advertencia.

**Se pueden configurar para detectar errores semánticos.**

```
# Ciclo for que avanza de dos en dos
for i in range (0, 100, 2):
  # Generar una excepción si el número es múltiplo de 10
  if i % 10 == 0:
    raise Exception("Lo lamentamos, no se admiten múltiplos de diez.")
  print(i)
```

### Uso combinado de while y try/except para controlar los datos ingresados por consola

```
def verificar_entrada_numero_telefonico ():
  '''Función que verifica que se ingrese un número telefónico '''
  while True:
    try:
      numero = input("Por favor ingrese el número de su teléfono sin dejar espacios en blanco ni usar caracteres especiales:")
      if len(numero) != 10:
        print("Recuerde que el número telefónico se compone de diez dígitos")
      elif numero.isnumeric():
        print("El número ingresado es válido")
        break
      else:
        numero = int(numero)
    except Exception:
      print("Recuerde no usar letras")
      print("Recuerde no dejar espacios en blanco")
      print("Recuerde no usar caracteres especiales")
  return numero

tel_usuario = verificar_entrada_numero_telefonico()
print(f"El número del usuario es: {tel_usuario}")
```
