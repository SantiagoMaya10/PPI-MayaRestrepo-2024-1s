```
# Escriba aquí su código

# Importar las librerías necesarias

# Se importa MatplotLib
import matplotlib
# Se importa NumPy
import numpy as np
# Se importa Pandas
import pandas as pd
# Se importa pyplot
from matplotlib import pyplot as plt
# Se importa seaborn
import seaborn as sns



# Conocer la versiones de las librerías instaladas
print('MatplotLib version:', matplotlib.__version__)
print('NumPy version:', np.__version__)
print('Pandas version:', pd.__version__)


# Guardar en la variable 'ruta' la url del dataset
ruta = "https://docs.google.com/spreadsheets/d/1zbYYyd_LjQYsLrnEu9HxNU2o7KgForKqmCmU-RXiVV4/pub?gid=1119984286&single=true&output=csv"

# Cargar el dataset a partir de la ruta establecida
data = pd.read_csv(ruta)


# Verificar la lectura del dataset

print(data.head())
print(data.dtypes)
```

## Ejercicio 00

Seleccione dos columnas numéricas del dataset y haga un gráfico tipo **plot**. Coloque título al gráfico y etiquete los dos ejes.

```
# Escriba su código aquí

# Se reasigna el dataset para que los cambios no se propaguen en los demás ejercicios
data_00 = data

# Limpiar la columna 'Price' quitando los signos de dólar y comas, strings
# con valor Not Priced y convirtiendo a numérico
data_00['Price'] = data_00['Price'].replace({'\$': '', ',': '',
                                            'Not Priced': np.nan}, regex=True)
data_00['Price'] = pd.to_numeric(data_00['Price'], errors = 'coerce')

# Limpiar la columna 'Mileage' quitando los strings y convirtiendo a numérico
data_00['Mileage'] = pd.to_numeric(data_00['Mileage'], errors='coerce')

# Eliminar filas con valores faltantes en 'Mileage'
data_00 = data_00.dropna(subset = ['Mileage'])

# Eliminar filas con valores faltantes en 'Price'
data_00 = data_00.dropna(subset = ['Price'])

# Ordenar el DataFrame por la columna 'Price'
data_00 = data_00.sort_values(by = 'Mileage')

# Seleccionar las columnas numéricas ('Mileage' y 'Price')
x_00 = data_00['Mileage']
y_00 = data_00['Price']

# Crear el gráfico tipo plot
plt.figure(figsize = (14, 8))
plt.plot(x_00, y_00, marker = 'o', color = 'b')

# Colocar título al gráfico y etiquetas en los ejes
plt.title('Relación entre el Kilometraje y el Precio de los Vehículos')
plt.xlabel('Kilometraje (Millas)')
plt.ylabel('Precio ($USD)')

# Mostrar el gráfico
plt.show()
fig_plot = plt.gcf()
```

## Ejercicio 01

Seleccione cinco columnas numéricas del dataset y haga un gráfico tipo **scatter**. Coloque título al gráfico y etiquete los dos ejes.

```
# Escriba su código aquí
plt.clf()

data_01 = data

# Convertir columnas a numéricas y eliminar filas con valores faltantes
data_01['Year'] = pd.to_numeric(data_01['Year'], errors='coerce')
data_01['Mileage'] = pd.to_numeric(data_01['Mileage'], errors='coerce')
data_01['Interior_Design_Rating'] = pd.to_numeric(data_01['Interior_Design_Rating']
                                                  , errors='coerce')
data_01['Price'] = pd.to_numeric(data_01['Price'], errors='coerce')
data_01['Comfort_Rating'] = pd.to_numeric(data['Comfort_Rating'], errors='coerce')
data_01 = data_01.dropna(subset=['Year', 'Mileage', 'Interior_Design_Rating',
                                 'Price', 'Comfort_Rating'])

# Seleccionar las columnas numéricas
x1_01 = data_01['Year']
x2_01 = data_01['Mileage']
x3_01 = data_01['Interior_Design_Rating']
y1_01 = data_01['Price']
y2_01 = data_01['Comfort_Rating']

# Crear el gráfico tipo scatter
plt.figure(figsize=(14, 8))
plt.scatter(x1_01, y1_01, label='Año vs. Precio', marker='x', color='b')
plt.scatter(x2_01, y1_01, label='Kilometraje vs. Precio', marker='o',
            color='r')
plt.scatter(x3_01, y1_01, label='Interior_Design_Rating vs. Precio',
            marker='s', color='g')
plt.scatter(x1_01, y2_01, label='Año vs. Comfort_Rating', marker='d', color='c')
plt.scatter(x3_01, y2_01, label='Interior_Design_Rating vs. Comfort_Rating',
            marker='^', color='m')

# Colocar título al gráfico y etiquetas en los ejes
plt.title('Gráfico de Dispersión de Cinco Columnas Numéricas')
plt.xlabel('Año, Kilometraje, Interior_Design_Rating')
plt.ylabel('Precio, Comfort_Rating')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()

fig_scatter = plt.gcf()
```

## Ejercicio 02

Seleccione cuatro columnas numéricas del dataset y haga un gráfico tipo **stackplot** (apilable). Coloque título al gráfico, etiquete los dos ejes e incluya una leyenda dentro del gráfico.

```
# Escriba su código aquí

plt.clf()

data_02 = data

# Reemplazar "Not Priced" por NaN y limpiar las columnas relevantes
data_02['Price'] = data_02['Price'].replace({'\$': '', ',': '',
                                             'Not Priced': np.nan},
                                            regex=True).astype(float)

# 2. Convertir valores no numéricos en las columnas de calificaciones a NaN
columns_to_convert = ['Comfort_Rating', 'Interior_Design_Rating',
                      'Performance_Rating', 'Reliability_Rating']
data_02[columns_to_convert] = data_02[columns_to_convert].apply(pd.to_numeric,
                                                                errors='coerce')

# 3. Eliminar filas con valores faltantes en las columnas seleccionadas
data_02 = data_02.dropna(subset=columns_to_convert)

# Seleccionar cuatro columnas numéricas
columns = ['Price', 'Comfort_Rating', 'Interior_Design_Rating',
           'Performance_Rating',
            'Reliability_Rating']

# Ordenar los datos por la primera columna seleccionada para el gráfico
data_02 = data_02.sort_values(by=columns[0])

# Preparar datos para el stackplot
x_02 = data_02[columns[0]]
y_02 = [data_02[col] for col in columns[1:]]

# Crear el gráfico tipo stackplot
plt.figure(figsize=(12, 8))
plt.stackplot(x_02, y_02, labels=columns[1:], alpha=0.7)

# Colocar título al gráfico y etiquetas en los ejes
plt.title('Stackplot de ratings de los compradores vs precio')
plt.xlabel(columns[0])
plt.ylabel('Ratings')

# Incluir una leyenda dentro del gráfico
plt.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
fig_stackplot = plt.gcf()
```

## Ejercicio 08

Seleccione las columnas que considere pertinentes del dataset y haga un mapa de calor (**heatmap**). Coloque título al gráfico y etiquete los dos ejes.

```
# Escriba su código aquí

# Seleccionar columnas de interés para el heatmap
columns_of_interest = [
    'Consumer_Rating', 'Comfort_Rating', 'Performance_Rating',
    'Value_For_Money_Rating', 'Exterior_Styling_Rating', 'Reliability_Rating'
]

# Filtrar el DataFrame para que contenga solo las columnas seleccionadas
filtered_data = data[columns_of_interest]

# Calcular la matriz de correlación
correlation_matrix = filtered_data.corr()

# Crear el heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlación de Calificaciones de Autos')
plt.show()
```

## Ejercicio 09

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico en 3D. Coloque título al gráfico y etiquete los ejes.

```
# Escriba su código aquí

data_09 = data

# Reemplazar "Not Priced" por NaN y limpiar la columna 'Price'
data_09['Price'] = data_09['Price'].replace({'\$': '', ',': '',
                        'Not Priced': None}, regex=True).astype(float)

# Seleccionar columnas de interés
x = data['Comfort_Rating']
y = data['Performance_Rating']
z = data['Price']

# Crear el gráfico 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Crear el gráfico de dispersión 3D
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# Configurar las etiquetas de los ejes
ax.set_xlabel('Comfort Rating')
ax.set_ylabel('Performance Rating')
ax.set_zlabel('Price')

# Agregar una barra de color
cbar = plt.colorbar(scatter, ax=ax, shrink=0.5, aspect=5)
cbar.set_label('Price')

plt.title('Gráfico 3D de Calificaciones de Autos')
plt.show()
```
